```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df=df.set_index('normalized_name')
df=df.drop(['Unnamed: 0'], axis=1)
df
df = df.drop(['bloom energy' , 'sigmacare','fisker automotive'])

```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

```python
df['funding_velocity'].max()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df=df.drop(['funding_velocity'], axis=1)
```

```python
df_majority = df[df.success==0]
df_minority = df[df.success==1]
df_minority['success'].value_counts()
```

```python
df_majority_downsampled = df_majority.sample(1546)
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
X = df_downsampled.loc[:, df_downsampled.columns != 'success']
y = df_downsampled.loc[:, df_downsampled.columns == 'success']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
           y, test_size=0.30, 
            random_state=101)
```

```python
X_train
```

```python
import shap
shap.initjs()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
#create an instance and fit the model 
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
Predictions = logmodel.predict(X_test)
```

```python
explainer = shap.KernelExplainer(logmodel.predict_proba, X_train)
shap_values = explainer.shap_values(X_test)
```

```python
columns_ = df_downsampled.iloc[:, df_downsampled.columns != 'success'].columns
```

```python
shap.summary_plot(shap_values, features=X_train, feature_names=columns_)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,Predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, Predictions))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, Predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
